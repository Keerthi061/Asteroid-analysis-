```python
#Example 2
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import urllib.request

# download data from Minor Planet Center
url = "https://minorplanetcenter.net/iau/MPCORB/MPCORB.DAT"
urllib.request.urlretrieve(url, "MPCORB.DAT")

# read data into a Pandas dataframe
cols = ['diameter']
data = pd.read_fwf("MPCORB.DAT", skiprows=45, usecols=[10], names=cols)

# convert diameter to decimal format
data['diameter'] = pd.to_numeric(data['diameter'], errors='coerce')

# filter out invalid data
data = data.dropna()

# create a histogram of asteroid diameters
plt.hist(data['diameter'], bins=np.logspace(np.log10(1), np.log10(1000), 20), alpha=0.5, color='b')
plt.xscale('log')
plt.xlabel('Estimated Diameter (km)')
plt.ylabel('Frequency')
plt.title('Histogram of Estimated Diameter of Asteroids')
plt.show()

# create a box plot of asteroid diameters
plt.boxplot(data['diameter'], vert=False)
plt.xscale('log')
plt.xlabel('Estimated Diameter (km)')
plt.title('Box Plot of Estimated Diameter of Asteroids')
plt.show()
```

```python
df1=pd.read_csv('/astdata1.csv')
df1
```

```python
# filter out invalid data and select relevant columns
df1= df1[df1['dist'].notnull()]
data = df1[['des', 'dist', 'v_rel']]

# convert distance and velocity to decimal format
data['dist'] = pd.to_numeric(data['dist'], errors='coerce')
data['v_rel'] = pd.to_numeric(data['v_rel'], errors='coerce')

# filter out invalid data
data = data.dropna()

# calculate mean and standard deviation of relative velocities
mean_velocity = np.mean(data['v_rel'])
std_velocity = np.std(data['v_rel'])

# print results
print(f"Mean relative velocity: {mean_velocity:.2f} km/s")
print(f"Standard deviation of relative velocities: {std_velocity:.2f} km/s")
```

```python
average_speed_jetliner = 2.0206208 # km/s
if mean_velocity > average_speed_jetliner:
    print("Asteroids during close approach are faster than a commercial jetliner.")
else:
    print("Asteroids during close approach are slower than a commercial jetliner.")

if std_velocity > 0.1:
    print("The relative velocities of asteroids during close approach are highly variable.")
else:
    print("The relative velocities of asteroids during close approach are relatively stable.")
```

```python
sns.relplot(data=data, x='dist', y='v_rel',aspect=1.61)
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# read data from CSV file
#data = pd.read_csv("asteroid_data.csv")

# set the value to compare to
compare_value = 2.02
colors = np.where(data["v_rel"] > compare_value, 'g', 'b')
# create a scatter plot
plt.scatter(data['dist'], data['v_rel'],c=colors)
plt.axhline(y=compare_value, color="r", linestyle="-")
plt.xlabel("dist")
plt.ylabel("velocity")
plt.title("Diameter of asteroids compared to " + str(compare_value) + " km")
plt.show()
```

```python
sns.relplot(data=data, x='dist', y='v_rel',hue=colors,aspect=1.61)
plt.show()
```

```python
import seaborn as sns
import pandas as pd
import numpy as np

# read data from CSV file
#data = pd.read_csv("asteroid_data.csv")

# set the value to compare to
compare_value = 3.12928

# create a list of colors based on whether each point is above or below the given value
colors = np.where(data["v_rel"] > compare_value, 'Fast', 'Slow')

# create a scatter plot with colored points
sns.scatterplot(data=data, x="dist", y="v_rel", hue=colors)
plt.axhline(y=compare_value, color="g", linestyle="--")
plt.xlabel("Closest Approach Distance")
plt.ylabel("V_rel")
plt.title("Relative Velocity of asteroids compared to Jetplane (3.12928 km/s) upon their closest approach")
plt.show()
```

```python
import numpy as np


# get the unique values and their counts
unique, counts = np.unique(colors, return_counts=True)

# create a frequency table as a dictionary
freq_table = dict(zip(unique, counts))

# print the frequency table
print(freq_table)
```

```python
import matplotlib.pyplot as plt


# create lists for x and y values
x_values = list(freq_table.keys())
y_values = list(freq_table.values())

# plot the bar graph
plt.bar(x_values, y_values)

# add labels and title
plt.xlabel('Relative Speed of asteroids')
plt.ylabel('Count')
plt.title('Relative Velocity of Asteroids with respect to NASA''s X-43A(3.12928 Km/s)')

# show the plot
plt.show()
```
