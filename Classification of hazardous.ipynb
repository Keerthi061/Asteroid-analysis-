```python
from sklearn.preprocessing import Normalizer
from sklearn.impute import SimpleImputer

categ = df1.columns
imputer = SimpleImputer()
df1_imputed = pd.DataFrame(imputer.fit_transform(df1[categ]), columns=categ)
transformer = Normalizer().fit(df1_imputed[categ])
X = transformer.transform(df1_imputed[categ])
y = df1[['Hazardous']]
y = np.array(y)
```

```python
from sklearn.model_selection import train_test_split
X_tr, X_te, y_tr, y_t = train_test_split(X, y, random_state = 0 , test_size = 0.3)
```

```python
from sklearn.metrics import accuracy_score,explained_variance_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import KFold
from sklearn.svm import SVC
```

```python
# KNN Model
def model_KNN(X_tr, y_tr, X_te, y_te):
    KNN = KNeighborsClassifier(n_neighbors = 3)
    KNN.fit(X_tr, y_tr)
    KNN_pred = KNN.predict(X_te)
    Acc_KNN = round(accuracy_score(KNN_pred, y_te) * 100, 2)
    return Acc_KNN

# RandomForest Model
def model_RFS(X_tr, y_tr, X_te, y_te):
    RF = RandomForestClassifier()
    RF.fit(X_tr, y_tr)
    RF_pred = RF.predict(X_te)
    Acc_RF = round(accuracy_score(RF_pred, y_te) * 100, 2)
    return Acc_RF

# SVM Model
def model_SVM(X_tr, y_tr, X_te, y_te):
    SVM = SVC(kernel='linear', C = 1.0)
    SVM.fit(X_tr, y_tr)
    SVM_pred = SVM.predict(X_te)
    Acc_SVM = round(accuracy_score(SVM_pred, y_te) * 100, 2)
    return Acc_SVM

# Gaussian Naive Bayes Model
def model_Gaussian_NB(X_tr, y_tr, X_te, y_te):
    GNB = GaussianNB()
    GNB.fit(X_tr, y_tr)
    GNB_pred = GNB.predict(X_te)
    Acc_GNB = round(accuracy_score(GNB_pred,y_te) * 100, 2)
    return Acc_GNB
```

```python
def K_Fold_Validation(K):
    kf = KFold(n_splits=K,random_state=None)
    acc_RFS, acc_KNN, acc_SVM, acc_GNB = [], [], [], []
    i = 1
    for train_index,test_index in kf.split(X):
        print("K FOLD {0} ITERATION:".format(i))
        X_train,X_test = X[train_index],X[test_index]
        y_train,y_test = y[train_index],y[test_index]
        y_train = y_train.ravel()
        y_test = y_test.ravel()
        res_RFS = model_RFS(X_train,y_train,X_test,y_test)
        res_KNN = model_KNN(X_train,y_train,X_test,y_test)
        res_SVM = model_SVM(X_train,y_train,X_test,y_test)
        res_GNB = model_Gaussian_NB(X_train,y_train,X_test,y_test)
        print_res(res_RFS,res_SVM,res_GNB,res_KNN)
        i+=1
        print()
    return acc_RFS, acc_KNN, acc_SVM, acc_GNB
```

```python
def print_res(res_RFS, res_SVM, res_GNB, res_KNN):
    print("Results of Random Forest Classifier:", res_RFS)
    print("Results of Support Vector Machine Classifier:", res_SVM)
    print("Results of Gaussian Naive Bayes Classifier:", res_GNB)
    print("Results of K-Nearest Neighbor Classifier:", res_KNN)
```

```python
K_Fold_Validation(3)
```

```python
acc_RFS, acc_KNN, acc_SVM, acc_GNB = K_Fold_Validation(3)
plt.bar(['RFS', 'KNN', 'SVM', 'GNB'], [np.mean(acc_RFS), np.mean(acc_KNN), np.mean(acc_SVM), np.mean(acc_GNB)])
plt.ylim([0, 100])
plt.title('K-Fold Validation Results')
plt.xlabel('Classifier')
plt.ylabel('Accuracy (%)')
plt.show()
```
