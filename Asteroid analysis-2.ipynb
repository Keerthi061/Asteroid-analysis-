```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('C:\\Users\\KEERTHI\\OneDrive\\Documents\\Preprocessing\\asteroid_final.csv')
df
```

```python
df = df.drop_duplicates()
df
```

```python
df=df.dropna(how='any',axis=0)
df
```

```python
orbital_params = ['Semi_Major_Axis', 'Eccentricity']
```

```python
orbital_df = df[orbital_params]
```

```python
x = orbital_df['Semi_Major_Axis']
y = orbital_df['Eccentricity']
kde = sns.kdeplot(x=x, y=y, cmap="Reds", fill=True, shade_lowest=False, bw_adjust=1)
kde.set_xlabel("Semimajor Axis")
kde.set_ylabel("Eccentricity")
kde.set_title("Asteroid Distribution in Orbital Space")
plt.show()
```

```python
sns.set_style('whitegrid')
m = sns.jointplot(x=x, y=y, kind='kde', cmap='Reds', shade=True, thresh=0.05, bw_adjust=1, fill=True)
m.ax_joint.set_title("Asteroid Distribution in Orbital Space")
# Show the plot
plt.show()
```

```python
variables = ['absolute_magnitude.y', 'Semi_Major_Axis', 'Eccentricity', 'Inclination', 'Aphelion_Dist']

# Calculate the correlation matrix
corr_matrix = df[variables].corr()

# Create a heatmap of the correlation matrix
sns.set_style('whitegrid')
sns.heatmap(corr_matrix, cmap='YlGnBu', annot=True)

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the asteroid data into a Pandas DataFrame
asteroids_df = pd.read_csv("C:\\Users\\adith\\astdata1.csv")

# Convert the Closest_Approach_Date column to a datetime object
asteroids_df['cd'] = pd.to_datetime(asteroids_df['cd'])

# Set the Closest_Approach_Date column as the index
asteroids_df.set_index('cd', inplace=True)

# Group the data by year and count the number of rows in each group
yearly_asteroids = asteroids_df.resample('Y').count()

# Plot the yearly asteroid counts
plt.plot(yearly_asteroids.index.year, yearly_asteroids['des'])
plt.xlabel('Year')
plt.ylabel('Number of asteroids')
plt.title('Number of asteroids closest to Earth per year')
plt.show()
```
